```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import quandl
import datetime as datetime
import yfinance as yf
from yahoofinancials import YahooFinancials
from pytrends import dailydata
from pytrends.request import TrendReq
```

```python
#data from blockchain.com using qunadl API
quandl.ApiConfig.api_key = '-7D4_v-SXngEYdwmFnfR' # get the API key to get datasets from quandl bitfinex database
df_price = quandl.get('BCHAIN/MKPRU') #bitcoin daily closing price
df_USD_ex_volume = quandl.get('BCHAIN/TRVOU') # daily exchange trade volume in USD
df_dif = quandl.get('BCHAIN/DIFF') # daily measure of difficutly of mining a block 
df_hash = quandl.get('BCHAIN/HRATE') #daily measure of hashrate
df_trasact_volume = quandl.get('BCHAIN/ETRAV') #daily number of transactional volume
df_transact_perblock = quandl.get('BCHAIN/NTRBL') # daily number of average transactions per block
df_transaction_no = quandl.get('BCHAIN/TRFUS') #daily number of total transactions
df_market_capitalization = quandl.get('BCHAIN/MKTCP') #bitcoin market capitalization by day
```

```python
dfl = [df_price, df_USD_ex_volume, df_dif, df_hash, df_trasact_volume, df_transact_perblock, df_transaction_no, df_market_capitalization]
#concatenate the different dataframes with an outerjoin, NaN values will be initial so should be filled with 0
from functools import reduce
df = reduce(lambda  left,right: pd.merge(left,right,on=['Date'], how='outer'), dfl).fillna(0)
#Rename the columns
df.columns = ['BTC_Price_USD', 'Trade_Vol_USD', 'Difficulty', 'HashRate', 'Transaction_Vol', 'Transactions_per_block', 'Total_Transactions','Market_Capitalization_USD']

#drop todays date because some columns will be null and will affect our analysis
df.drop(datetime.date.today(),axis = 0)
```

```python
#Now adding the tech stocks accounted in NASDAQ 100
df_NASDAQ= yf.download('^OEX', start='2009-01-02',end=datetime.date.today() - datetime.timedelta(days=1), 
                       progress=False) #NASDAQ-100 data for tech stocks
df_NASDAQ= df_NASDAQ.drop(['Open','High','Low','Adj Close','Volume'], 1)
df_NASDAQ = df_NASDAQ.rename(columns={'Close': 'NASDAQ-100',})
df_NASDAQ.head()
```

```python
#Now for S&P 500
df_SandP500= yf.download('^GSPC', start='2009-01-02',end=datetime.date.today() - datetime.timedelta(days=1), 
                       progress=False) # S&P 500 data
df_SandP500= df_SandP500.drop(['Open','High','Low','Adj Close','Volume'], 1)
df_SandP500 = df_SandP500.rename(columns={'Close': 'S&P 500',})
df_SandP500.tail()
```

```python
#Now accounting for gold and silver, using the philadelphia gold and silver index
df_gold= yf.download('^XAU', start='2009-01-02',end=datetime.date.today() - datetime.timedelta(days=1), 
                       progress=False) #XAU index
df_gold= df_gold.drop(['Open','High','Low','Adj Close','Volume'], 1)
df_gold = df_gold.rename(columns={'Close': 'Gold & Silver',})
df_gold.tail()
```

```python
# Now accounting for brent Oil price
df_oil= yf.download('CL=F', start='2009-01-02',end=datetime.date.today() - datetime.timedelta(days=1), 
                       progress=False) #XAU index
df_oil= df_oil.drop(['Open','High','Low','Adj Close','Volume'], 1)
df_oil = df_oil.rename(columns={'Close': 'Oil',})
df_oil.tail()
```

```python
#Now to get google search index data
df_interest = dailydata.get_daily_data('bitcoin', 2009, 1, 2022, 3)
```

```python
df_interest = df_interest.drop(['bitcoin_unscaled','bitcoin_monthly','isPartial','scale'], 1)
```

```python
df_interest = df_interest.rename(columns={'bitcoin': 'Interest_Index'})
df_interest.index.names = ['Date']
df_interest
```

```python
# fill forward and backward to fill missing data of stock indexes on weekends 
dfl2 = [df, df_NASDAQ, df_SandP500, df_gold, df_oil, df_interest] 
df_final = reduce(lambda  left,right: pd.merge(left,right,on=['Date'], how='outer'), dfl2).fillna(method ='bfill').fillna(method='ffill')
df_final.tail()
```
